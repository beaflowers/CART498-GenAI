```python
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"

from diffusers import DiffusionPipeline

model_id = "Tinylilpoptart/riley"
pipe = DiffusionPipeline.from_pretrained(model_id,
                                         dtype=torch.bfloat16, device_map="cuda")
```

```python
prompt = "There is a gas station along a highway, packed with cars. A giant woman is 1000 feet tall, looming over the gas station with her arms outstretched. She towers over everything. The image is brilliantly colored with a blue sky."
image = pipe(prompt).images[0]
image
image.save("output_image.png")
```
